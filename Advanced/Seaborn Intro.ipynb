```python
import numpy as np
import pandas as pd
import ast
import seaborn as sns
import matplotlib.pyplot as plt
from datasets import load_dataset
from matplotlib.ticker import FuncFormatter
```

```python
# Load the dataset from the CSV file
dataset = load_dataset('csv', data_files=r'C:\Users\sami.laham\Documents\Sam\Data Analysis Training\Luke Barousse\CSV files\data_jobs.csv')

# Convert the dataset to a pandas DataFrame
df = dataset['train'].to_pandas()

# Data Cleaning: Convert the 'job_posted_date' to datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
# Show the first few rows of the DataFrame to check the results
df
```

```python
df_DA_Spain = df [(df['job_title_short'] == 'Data Analyst') & (df ['job_country'] == 'Spain')].copy()
df_DA_Spain = df_DA_Spain.dropna(subset = ['salary_year_avg'])

df_DA_Spain
```

```python
df_DA_Spain = df_DA_Spain.explode('job_skills')

df_DA_top_pay = df_DA_Spain.groupby ('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values( by = 'median', ascending = False)
df_DA_top_pay = df_DA_top_pay.head(10)

df_DA_skills = df_DA_Spain.groupby ('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values( by = 'count', ascending = False)
df_DA_skills = df_DA_skills.head(10).sort_values ( by = 'median', ascending= False)

fig, ax = plt.subplots (2, 1)

df_DA_top_pay[::-1].plot (kind = 'barh', y = 'median', ax = ax [0], legend = False)
ax[0].set_title ('Top 10 Highest Paid Skills for Data Analysts')
ax[0].set_ylabel('')
ax[0].set_xlabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'â¬{int(x / 1000)}K'))

df_DA_skills[::-1].plot (kind = 'barh', y = 'median', ax = ax [1], legend = False)
ax[1].set_title ('Top 10 Most In-Demand Skills for Data Analysts')
ax[1].set_ylabel ('')
ax[1].set_xlabel ('Median Salary (EUR)')
ax[1].set_xlim (ax[0].get_xlim())
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'â¬{int(x / 1000)}K'))

fig.tight_layout()
```
