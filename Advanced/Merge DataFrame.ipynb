```python
df_Spain_software_pivot = pd.read_csv('C:\\Users\\sami.laham\\Documents\\Sam\\Data Analysis Training\\Luke Barousse\\CSV files\\software_jobs.csv', index_col = 'job_posted_month')
df_Spain_software_pivot
```

```python
df_Spain_merged = df_Spain_pivot.merge(df_Spain_software_pivot, on = "job_posted_month")
```

```python
top_5 = df_Spain_merged
.sum()
.sort_values(ascending=False)
.head()
.index
.to_list()

top_5
```

```python
df_Spain_merged[top_5].plot(kind = 'line')
plt.title ('Monthly Job Postings for Top Tech Jobs in Spain')
plt.xlabel ('2023')
plt.ylabel ('Job Count')
plt.ylim(0, 25000)
plt.legend()
plt.show()
```
