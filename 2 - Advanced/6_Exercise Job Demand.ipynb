```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
# Loading Data
dataset = load_dataset('csv', data_files=r'C:\Users\lahms\OneDrive\Documents\Sam - Data Analyst\data_jobs.csv')
df = dataset ['train'].to_pandas()

# Data Cleaning

df ['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
print (dataset)
```

```python
df_Spain = df[df['job_country'] == 'Spain']
df_Spain
```

```python
df_Spain['job_posted_month'] = df_Spain ['job_posted_date'].dt.strftime('%B')
df_Spain
```

```python
df_Spain
```

```python
df_Spain.pivot_table(index= 'job_posted_month', columns = 'job_title_short', aggfunc = 'size')
```

```python
df_Spain_pivot = df_Spain.pivot_table(index= 'job_posted_month', columns = 'job_title_short', aggfunc = 'size')
df_Spain_pivot
```

```python
df_Spain_pivot.reset_index(inplace=True)
```

```python
df_Spain_pivot ['month_num'] = pd.to_datetime (df_Spain_pivot ['job_posted_month'], format = '%B').dt.month
df_Spain_pivot ['month_num']
```

```python
df_Spain_pivot
```

```python
df_Spain_pivot.sort_values('month_num', inplace = True)
```

```python
df_Spain_pivot.set_index('job_posted_month', inplace=True)
```

```python
df_Spain_pivot.drop(columns = 'month_num', inplace = True)
```
