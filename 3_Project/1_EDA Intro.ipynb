```python
df_plot = df_DA_Spain ['job_location'].value_counts().head(10).to_frame()
sns.set_theme (style = 'ticks')
sns.barplot(data =df_plot, x = 'count', y = 'job_location', hue = 'count', palette = 'flare', legend = False)
sns.despine()
plt.title ('Counts of Data Analyst Opportunies Per Job Location in Spain')
plt.xlabel('Number of Opportunies')
plt.ylabel('')
plt.show()
```

```python
dict_column = { 
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
}

fig, ax = plt.subplots(1, 2)
fig.set_size_inches((12, 5))
for i, (column, title) in enumerate (dict_column.items()):
    ax[i].pie(df_DA_Spain[column].value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

fig.tight_layout()
plt.show()
```
